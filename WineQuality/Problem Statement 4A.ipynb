```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
from numpy.random import randint,randn
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
wineset_frame_org = pd.read_csv('winequality-red.csv', sep = ';')
```

```python
wineset_frame_org.loc[459]
```

```python
limited_frame = DataFrame(wineset_frame_org,columns=['quality', 'total sulfur dioxide'])
```

```python
limited_frame
```

```python
limited_frame.rename(columns={'total sulfur dioxide':'Total_so2'},inplace = True)
```

```python
sorted_so2 = DataFrame(limited_frame.sort_values(["quality","Total_so2"]))
```

```python
sorted_so2.head(30)
```

```python
del sorted_so2["quality_sorted"]
```

```python
grp1 = sorted_so2['Total_so2'].groupby(sorted_so2['quality'])
```

```python
grp1.max()
```

THE MAXIMUM CONTENT OF SO2 IS FOUND IN WINE OF QUALITY 7. THEREFORE IT IS DANGEROUX

```python
ranking = DataFrame(grp1.max().sort_values())
```

```python
ranking['Rank'] = [1,2,3,4,5,6]
```

```python
ranking
```

```python
quality_list = sorted_so2.quality.tolist()
so2_list = sorted_so2.Total_so2.tolist()
plt.bar(quality_list,so2_list,color = 'red', align = 'center')
plt.xlabel("Quality")
plt.ylabel("Total Sulphur Dioxide")
plt.show()
```
