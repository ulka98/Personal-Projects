#   Problem Statement 4 :3. Find a correlation between pH and Quality. 
# Fact: Winemakers use pH as a way to measure ripeness in relation to acidity.Low pH wines will taste tart and crisp, while higher pH wines are more susceptible to bacterial growth.Most wine pH's fall around 3 or 4; about 3.0 to 3.4 is desirable for white wines, while about 3.3 to 3.6 is best for reds.

# Expected analysis: Higher quality wines should contain pH between 3 and 3.4

```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
from numpy.random import randint,randn
```

```python
wineset_frame_org = pd.read_csv('winequality-red.csv', sep = ';')
```

```python
wineset_frame_org.head(10)
```

```python
wineset_frame_org['pH'].corr(wineset_frame_org['quality'])
```

```python
wineset_frame_org['pH'].corr(wineset_frame_org['residual sugar'],method='pearson')
```

Since the corr value isn't near to -1 or +1, both the factors are necessary for analysis and we can't ignore both. Also both of the relations above has negative pearson's coefficient, therefore, rescective variables posses inverse relationship 

```python
qualities = wineset_frame_org.quality.tolist()
pH_values = wineset_frame_org.pH.tolist()
plt.scatter(pH_values,qualities,c='cyan',marker = 's',s =20)
plt.xlabel('pH')
plt.ylabel('quality')
plt.grid()
plt.show()
```
