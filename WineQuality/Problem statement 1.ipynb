# Problem Statement 1: Plot Hist and Line graph for pH level for Quality 5 and Quality 8


```python
import numpy as numpy
import pandas as pd
from pandas import DataFrame, Series
from numpy.random import randint,randn
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
wineset_frame_org = pd.read_csv('winequality-red.csv', sep = ';')
```

```python
wineset_frame_org.head()
```

```python
limited_frame = DataFrame(wineset_frame_org,columns = ['pH','quality'])
```

```python
quality8_frame = limited_frame[limited_frame['quality'] == 8]
```

```python
quality8_frame
```

```python
quality5_frame = limited_frame[limited_frame['quality'] == 5]
```

```python
quality5_frame
```

```python
quality_5 = quality5_frame.quality.tolist()
quality_8 = quality8_frame.quality.tolist()
pH_5 = quality5_frame.pH.tolist()
pH_8 = quality8_frame.pH.tolist()
plt.plot(pH_5,quality_5,label='Quality 5',color = 'blue', marker = 'o')
plt.plot(pH_8,quality_8,label='Quality 8', color = 'orange', marker = 's')
plt.xlabel('pH level')
plt.ylabel('Quality')
plt.grid()
plt.legend()
plt.show()
```

```python
quality_5 = quality5_frame.quality.tolist()
quality_8 = quality8_frame.quality.tolist()
pH_5 = quality5_frame.pH.tolist()
pH_8 = quality8_frame.pH.tolist()
plt.hist(pH_5,bins = [2.8,2.9,3.0,3.1,3.2,3.3,3.4,3.5,3.6,3.7,3.8,3.9,4.0],color = 'green',rwidth=0.95)
plt.ylabel('Quality')
plt.xlabel('pH level')
plt.legend()
plt.show()
```
