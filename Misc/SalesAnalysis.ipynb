Count of the historical and active state, (we need only analyze the active state products):

```python
hist_count = sales[sales['File_Type']=='Historical']['SKU_number'].count()
actv_count = sales[sales['File_Type']=='Active']['SKU_number'].count()
print(hist_count,actv_count)
```

Frequency distribution of Marketing type

```python
sns.countplot('MarketingType',data=sales)
plt.show()
```

```python
col = sales[sales.columns[-6:]].columns
col
```

```python
fig, ax = plt.subplots(6,1,figsize=(10,30))
for i, col_val in enumerate(col):
    sns.distplot(sales[col_val],hist=True,color='red',ax=ax[i])
    ax[i].set_label('freq dist of '+ col_val)
    ax[i].set_xlabel(col_val)
    ax[i].set_ylabel('count')

plt.tight_layout()
plt.show()
    
```

```python
sns.pairplot(sales)
```

```python
sales_pairplot = sales.drop([
      'Order', 'File_Type','SKU_number','SoldFlag','MarketingType','ReleaseNumber','New_Release_Flag'
], axis=1)
sns.pairplot(sales_pairplot)
plt.show()
```

```python
sns.scatterplot(y='SoldFlag',x='ReleaseYear',data=sales,palette='bright',hue='MarketingType')
#plt.legend(custom, ['Yes', 'No'], loc='lower right')
plt.legend(loc='center')
plt.show()
```

```python
historical = sales[sales['File_Type']=='Historical']
```

```python
cond = sns.FacetGrid(data=historical, col='MarketingType')
cond.map(sns.scatterplot,'ReleaseYear','SoldFlag').add_legend()
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots()
for a in [historical['ReleaseYear'],historical['SoldFlag']]:
    sns.distplot(a, bins=range(1900, 2020, 10), ax=ax, kde=False)
ax.set_xlim([1960, 2020])
```
